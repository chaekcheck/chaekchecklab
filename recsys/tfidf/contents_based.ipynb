```python
import pandas as pd
import os
```

```python
# 폴더 내에 있는 여러개의 파일 불러오기

# CSV 파일이 있는 폴더 경로
folder_path = 'basic/'

# 폴더 내의 모든 파일 목록 가져오기
files = os.listdir(folder_path)

# CSV 파일을 하나의 데이터프레임으로 합치기
df_list = []

for file in files:
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        temp_df = pd.read_csv(file_path)
        df_list.append(temp_df)

# 하나로 합치기
combined_basic_df = pd.concat(df_list, ignore_index=True)

combined_basic_df
```

```python
###### full_name 칼럼을 append 하는 걸로 ########

def get_recommendations(title, cosine_sim=cosine_matrix, num_recommendations=10):
    # 선택한 책의 타이틀로부터 해당 책의 인덱스 가져오기
    if title not in title_to_index:
        raise ValueError(f"'{title}'는 데이터베이스에 없습니다.")
        
    idx = title_to_index[title]

    # 해당 책과 모든 책과의 유사도 가져오기
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도에 따라 책 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    unique_recommendations = []
    seen_books = set()  # 중복 제거를 위한 집합

    for sim_score in sim_scores:
        book_idx = sim_score[0]
        full_name = df_INNER_JOIN.loc[book_idx, 'full_name']
        book_url = df_INNER_JOIN.loc[book_idx, 'book_url']  # 책 URL 가져오기
        
        # 입력한 책과 같은 제목인 경우 제외
        if full_name == title:
            continue

        # 중복 확인: full_name이 같고 URL도 동일한 경우
        book_identifier = (full_name, book_url)
        if book_identifier in seen_books:
            continue  # 중복된 책은 건너뛰기

        seen_books.add(book_identifier)  # 새로운 책 추가
        unique_recommendations.append((book_idx, full_name))  

        # 추천할 책의 수가 num_recommendations에 도달하면 종료
        if len(unique_recommendations) == num_recommendations:
            break

    # 결과를 데이터프레임으로 변환
    result = pd.DataFrame(unique_recommendations, columns=['index', 'full_title'])
    return result.reset_index(drop=True)

```

```python
def get_start_book_details(input_title, df_INNER_JOIN):
    # 입력한 제목으로 시작하는 책 찾기
    matching_books = df_INNER_JOIN[df_INNER_JOIN['full_name'].str.startswith(input_title)]

    # 일치하는 책이 있는지 확인
    if matching_books.empty:
        return f"'{input_title}'로 시작하는 책이 데이터베이스에 없습니다."
    
    # 일치하는 책의 상세정보 반환
    return matching_books.reset_index(drop=True)
```

```python
def get_in_book_details(input_title, df_INNER_JOIN):
    # 입력한 제목이 포함된 책 찾기
    matching_books = df_INNER_JOIN[df_INNER_JOIN['full_name'].str.contains(input_title, na=False)]

    # 일치하는 책이 있는지 확인
    if matching_books.empty:
        return f"'{input_title}'가 포함된 책이 데이터베이스에 없습니다."
    
    # 일치하는 책의 상세정보 반환
    return matching_books.reset_index(drop=True)


input_title = '칭기즈칸' 
book_details = get_in_book_details(input_title, df_INNER_JOIN)

print(book_details) 
```
