```python
import cv2
from main import PororoOcr

# 1. 이미지 읽기
img_path = 'assets/images/im.jpg18_2.jpg'
image = cv2.imread(img_path)

# 2. 이미지 리사이즈
resized_image = cv2.resize(image, (1200, 400))

# 3. 전처리된 이미지를 임시 경로에 저장
processed_img_path = 'assets/images/resized_image.jpg'
cv2.imwrite(processed_img_path, resized_image)

# 4. OCR 실행
ocr = PororoOcr()
ocr.run_ocr(processed_img_path, debug=True)
```
