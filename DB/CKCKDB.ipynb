```python
conn = pymysql.connect(
    host=os.environ["a_host"],
    port = int(os.environ["a_port"]),
    database =os.environ["a_database"],
    user=os.environ["a_user"],
    password=os.environ["a_password"],
    charset=os.environ['charset']
)
```

```python
# tb_book 생성

mk_book_table = '''
CREATE TABLE IF NOT EXISTS tb_book (
id INTEGER UNSIGNED NOT NULL AUTO_INCREMENT,
title VARCHAR(500) NOT NULL,
full_title VARCHAR(500) NOT NULL,
author VARCHAR(500) NOT NULL,
translator VARCHAR(500) NULL,
etc_author VARCHAR(500) NULL,
author_list VARCHAR(500) NULL,
publish_date DATE NULL,
publisher VARCHAR(50) NULL,
introduction TEXT NULL,
isbn VARCHAR(20) NOT NULL,
page VARCHAR(10) NULL,
categories VARCHAR(500) NOT NULL,
url VARCHAR(100) NOT NULL,
PRIMARY KEY (id),
UNIQUE KEY unique_name (url)
)
'''

with conn.cursor() as cursor :
    cursor.execute(mk_book_table)
    conn.commit()
```

```python
mk_USER1_table = '''
CREATE TABLE IF NOT EXISTS tb_user_books (
id INTEGER UNSIGNED NOT NULL AUTO_INCREMENT,
book_id INTEGER UNSIGNED NOT NULL,
PRIMARY KEY (id),
FOREIGN KEY (book_id) REFERENCES tb_book(id)
)
'''

with conn.cursor() as cursor :
    cursor.execute(mk_USER1_table)
    conn.commit()
```

```python
require = ['title','full_name','저자', '역자', '기타', 'author_list','date','publisher','intro','isbn','pages', 'category', 'book_url' ]
insert_sql = """INSERT IGNORE INTO tb_book (title, full_title, author, translator, etc_author, author_list, publish_date, publisher, introduction, isbn, page, categories, url) VALUES 
(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);"""

batch_size = 2000

try:
    with conn.cursor() as cursor:
        for start_idx in range(0, len(df), batch_size):
            batch_data = [tuple(df.loc[idx, require]) for idx in range(start_idx, min(start_idx + batch_size, len(df)))]
            
            cursor.executemany(insert_sql, batch_data)
            print(f"{start_idx + len(batch_data)}개 데이터 삽입 완료")
        
        conn.commit()

except Exception as e:
    print(f"에러 발생: {e}")

finally:
    conn.close()
```
