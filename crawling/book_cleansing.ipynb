```python
import os
import pandas as pd

book_folder = r'C:\Users\white\Downloads\basic\basic'  
detail_folder = r'C:\Users\white\Downloads\detail\detail'  

# 결과 저장 폴더 설정
output_folder = r'C:\Users\white\Downloads\다운로드'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

def extract_number(filename):
    return filename.split('_')[-1].split('.')[0]

book_files = [f for f in os.listdir(book_folder) if f.startswith('basic')]
detail_files = [f for f in os.listdir(detail_folder) if f.startswith('detail')]

# 모든 파일 병합 및 처리
for book_file in book_files:
    number = extract_number(book_file)
    
    detail_file = f'detail_{number}.csv'
    
    if detail_file in detail_files:
        book_df = pd.read_csv(os.path.join(book_folder, book_file))
        detail_df = pd.read_csv(os.path.join(detail_folder, detail_file))
        
        merged_df = pd.merge(book_df, detail_df, on='book_url', how='inner')
        
    
        merged_df['category'] = number
        
        output_file = os.path.join(output_folder, f'merged_{number}.csv')
        merged_df.to_csv(output_file, index=False)
        print(f'{output_file} 저장 완료.')

print("모든 파일 병합 및 저장 완료.")
```

```python
import pandas as pd
import re

# 정보를 분리하여 각 컬럼에 저장하는 함수 정의
def extract_info(info_str):
    # 리스트 형태로 변환
    info_list = eval(info_str)  # 문자열을 리스트로 변환

    # 기본 값 설정
    date, pages, weight, thick = None, None, None, None
    
    # 날짜 추출 (XXXX년 XX월 XX일 형식)
    for info in info_list:
        if re.match(r"\d{4}년 \d{2}월 \d{2}일", info):
            date = info
        elif "쪽" in info:
            # 쪽, g, mm 정보 분리
            parts = info.split('|')
            if len(parts) >= 3:  # 길이 체크
                pages = parts[0].strip()  # '264쪽'
                weight = parts[1].strip()  # '525g'
                thick = parts[2].strip()  # '153*224*20mm'
            else:
                # parts의 길이가 부족할 경우 기본값 설정
                pages = parts[0].strip() if len(parts) > 0 else None
                weight = parts[1].strip() if len(parts) > 1 else None
                thick = None  # 두 번째 값이 없을 경우 None으로 설정

    return pd.Series([date, pages, weight, thick])

# book 데이터프레임에서 'item_infos' 컬럼에 함수 적용하여 새로운 컬럼 추가
book[['date', 'pages', 'weight', 'thick']] = book['item_infos'].apply(extract_info)

# 결과 확인
print(book.head())
```

```python
# 다른 칼럼이 모두 동일하나, cates, category, item_infos가 다른 경우 category 병합
book = book.groupby(['title', 'full_name', 'book_url', 'authors', 'publisher', 'author_list', 'intro', 'date', 'pages', 'weight',	'thick']).agg({
    'category': lambda x: ', '.join(sorted(set(map(str, x))))  # 정수형을 문자열로 변환하여 병합
}).reset_index()
book
```

```python
#동일한 제목이나 개정되어 여러 권 있는 경우 병합(가장 최근 책만 남겨둠)
```

```python
book['date'] = pd.to_datetime(book['date'], format='%Y년 %m월 %d일')

# 'title'이 동일한 행들에 대해 'date'를 기준으로 정렬 (최신 날짜가 위로 오게)
book = book.sort_values('date', ascending=False)

# 'title'을 기준으로 그룹화하고, 최신 'date'의 데이터를 유지
book_merged = book.groupby('title', as_index=False).agg({
    'full_name': 'first',  # 첫 번째 값 유지
    'book_url': 'first',   # 첫 번째 값 유지
    'authors': 'first',    # 첫 번째 값 유지
    'publisher': 'first',  # 첫 번째 값 유지
    'author_list': 'first',# 첫 번째 값 유지
    'intro': 'first',      # 첫 번째 값 유지
    'date': 'first',       # 최신 날짜 유지
    'pages': 'first',      # 첫 번째 값 유지
    'weight': 'first',     # 첫 번째 값 유지
    'thick': 'first',      # 첫 번째 값 유지
    'category': lambda x: ', '.join(sorted(set(map(str, x))))  # 중복 없이 category 병합
})
```
